```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
list(df)
```

```python
def convert_to_number(x):
    try:
        return int(x)
    except:
        return 0
df.revenue = df.revenue.apply(convert_to_number)
```

```python
df.sort_values('revenue', ascending=False, inplace=True, na_position='last')
df.index = range(len(df))
```

```python
required_columns = 'title Actors Directors Poster vote_count vote_average revenue overview'
movie_frame = df[required_columns.split()].head(100)
movie_frame['id'] = range(len(movie_frame))
```

```python
movie_frame.to_csv('filter_movies.csv', index=False)
```
